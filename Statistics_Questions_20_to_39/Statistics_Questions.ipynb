<h3>Q20. What do you mean by measure of central tendency and measure of central dispersion? How it can be calculated?</h3>
<p><b>Answer</b></p>

<h4>Measure of Central Tendency</h4>
measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set. The most common measures of central tendency are:

1. Mean: Add up all the values in the data set and divide by the number of values.<br>
2. Median: The middle value when the data set is ordered from smallest to largest. If the data set has an even number of observations, the median is the average of the two middle numbers.<br>
3. Mode: The value that appears most frequently in the data set. A data set may have one mode, more than one mode, or no mode at all.

<h4>Measure of Central Dispersion</h4>
A measure of central dispersion (or variability) describes the spread of data points around the central value. The most common measures of central dispersion are:

1. Range: The difference between the largest and smallest values in the data set.<br>
2. Variance: Also known as spread of the data. It is calculated as the sum of the square difference of data point and mean divided by n (in case population ) or n-1 (in case of sample). In case of population variance it denoted by sigma square and in case of sample data it denoted by S square.<br>
3. Standard Deviation:  The square root of the variance. It provides a measure of the average distance of each data point from the mean.<br>
4. Interquartile Range (IQR): Formula: IQR=Q3âˆ’Q1, The difference between the third quartile (Q3) and the first quartile (Q1). It measures the spread of the middle 50% of the data.
 


<h3>Q21. What do you mean by skewness. Explain its types. Use graph to show.</h3>
<p><b>Answer</b></p>
Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates whether the data points are more concentrated on one side of the mean than the other. Skewness can be positive, negative, or zero.

<h4>Types of Skewness</h4>

1. Right-Skewed Distribution 
2. Left-Skewed Distribution
3. Symmetrical Distribution


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Generate data for each type of skewness
np.random.seed(0)
size = 1000

# right skew or positive skew
data_pos_skew = np.random.exponential(scale=2, size=size)

# left skew or negative skew
data_neg_skew = -np.random.exponential(scale=2, size=size)

# normal distribution or zero skew
data_zero_skew = np.random.normal(loc=0, scale=1, size=size)


# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Positive Skewness
sns.histplot(data_pos_skew, kde=True, color='blue', ax=axs[0])
axs[0].set_title('Right Skew')

# Negative Skewness
sns.histplot(data_neg_skew, kde=True, color='green', ax=axs[1])
axs[1].set_title('Left Skew')

# Zero Skewness
sns.histplot(data_zero_skew, kde=True, color='red', ax=axs[2])
axs[2].set_title('Normal Skew')

# Show the plot
plt.tight_layout()
plt.show()


```

<p><b>Answer</b></p>


<h4>Probability Mass Function (PMF)</h4>
    <p>
        The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. It is used for discrete random variables, where the set of possible outcomes is countable.
    </p>
    <p>
        <strong>Formula:</strong> P(X = x)
    </p>
    <p>
        Here, P(X = x) represents the probability that the random variable (X) takes the value (x).
    </p>
    <p>
        For example, if (X) is a discrete random variable representing the roll of a fair six-sided die, the PMF is given by:
    </p>
    <pre>
        P(X = 1) = 1/6
        P(X = 2) = 1/6
        P(X = 3) = 1/6
        P(X = 4) = 1/6
        P(X = 5) = 1/6
        P(X = 6) = 1/6
    </pre>

<h4>Probability Density Function (PDF)</h4>
    <p>
        The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable to take on a particular value. Unlike the PMF, the value of the PDF at any given point does not represent the actual probability, but rather the density of the probability. The probability that a continuous random variable falls within a certain interval is obtained by integrating the PDF over that interval.
    </p>
    <p>
        <strong>Formula:</strong> f(x)
    </p>
    <p>
        Here, f(x)  represents the density of the random variable (X) at the value ( x).
    </p>
    <p>
        For example, if (X) is a continuous random variable representing the height of adult males, the PDF might look something like this:
    </p>
    <pre>
        f(x) = (1 / (Ïƒâˆš(2Ï€))) * exp(-0.5 * ((x - Î¼) / Ïƒ)^2)
    </pre>
    <p>
        where (Î¼) is the mean and ( Ïƒ ) is the standard deviation.
    </p>

<h4>Differences Between PMF and PDF</h4>
    <ul>
        <li>
            <strong>Type of Variable:</strong>
            <ul>
                <li>PMF: Used for discrete random variables.</li>
                <li>PDF: Used for continuous random variables.</li>
            </ul>
        </li>
        <li>
            <strong>Probability Representation:</strong>
            <ul>
                <li>PMF: Directly gives the probability that a random variable is exactly equal to a specific value.</li>
                <li>PDF: Represents the density of the probability at a specific value; the actual probability is obtained by integrating the PDF over an interval.</li>

            <ul>

<h3>Q23. What is correlation> Explain its type in details. What are the methods of determining-correlation.</h3>
<p><b>Answer</b></p>

Correlation is a statistical measure. It indicates the strength and direction of a linear relationship between two random variables. Correlation is often used in data analysis to determine how changes in one variable might changes in another variable.

Types of Correlation

1. Positive Correlation
    - When two variables move in the same direction. As one variable increases, the other also increases, and as one decreases, the other also decreases.
2. Negative Correlation
    - When two variables move in opposite directions. As one variable increases, the other decreases, and vice versa.
3. No Correlation  
    - When there is no apparent relationship between two variables. Changes in one variable do not predict changes in the other.

Methods:
- Pearson: - for linear relationships with normally distributed data.
- Spearman: - for ordinal data or non-linear relationships.
- Kendall: - for small samples or tied ranks.
- Point-Biseria: - for relationships between binary and continuous variables.

<h3>Q26. Find the most likely price at Delhi corresponding to the price of Rs. 70 at Agra from the following data: Coefficient of correlation between the prices of the two places +0.8.</h3>
<p><b>Answer</b></p>

```python
# Given values
r = 0.8

## Let assume some values.
mean_price_agra = 65
mean_price_delhi = 68
std_dev_agra = 8
std_dev_delhi = 6

# Calculate slope (b)
b = r * (std_dev_delhi / std_dev_agra)

# Calculate intercept (a)
a = mean_price_delhi - (b * mean_price_agra)

# Given price at Agra
price_agra = 70

# Predict price at Delhi
price_delhi = a + (b * price_agra)

price_delhi
```

<h3>Q28. What is Normal Distribution? What are the four Assumptions of Normal Distribution? Explain in details.</h3>
<p><b>Answer.</b></p>

Normal distribution, also known as Gaussian distribution, is a fundamental concept in statistics and probability theory. It describes a continuous probability distribution of a random variable where data clusters around a central mean value with a characteristic symmetric, bell-shaped curve. The shape of the curve is determined by two parameters: the mean (Î¼), which represents the center of the distribution, and the standard deviation (Ïƒ), which measures the spread or dispersion of the data points around the mean.

Four Assumptions of Normal Distribution:

1. Data Symmetry and Bell-Shaped Curve
    - The distribution is symmetric around the mean Î¼.
2. Mean, Median, and Mode Equality
    - The mean (Î¼), median, and mode of the distribution are equal.  
3. Exact Probability Tail Behavior   
    - The tails of the distribution asymptotically approach but never touch the horizontal axis (x-axis).
4. Constant Shape
    - The shape of the distribution is consistent regardless of the mean and standard deviation values.

<h3>Q34. What is the statistical hypothesis? Explain the errors in hypothesis testing. Explain the samples. What are Large Samples & Small Samples?</h3>
<p><b>Answer</b></p>

The statistical hypothesis is used to claim or conclude the population parameters such as a mean, mode, variance, proportion etc. We can conduct the hypothesis about the population parameters just taking and doing some calculation on sample data. There are two types of hypotheses in hypothesis testing:

1. Null Hypothesis (Hâ‚€): This hypothesis assumes that there is no significant difference or relationship between variables or no effect of a treatment. It represents the status quo or the default position.

2. Alternative Hypothesis (Hâ‚ or Hâ‚): This hypothesis states what we are testing for; it represents the opposite of the null hypothesis and suggests that there is a significant difference, effect, or relationship.

<h4>Errors in Hypothesis Testing</h4>

1. Type I Error (Î±): This occurs when we reject the null hypothesis when it is actually true. 
2. Type II Error (Î²): This occurs when we fail to reject the null hypothesis when it is actually false.

<h4>Samples</h4> 
Sample is a subset taken from a larger population. The purpose of sampling is to draw conclusions about the entire population based on the characteristics observed in the sample. 

<h4>Large Samples & Small Samples</h4>

- Large Samples: A large sample generally refers to a sample size that is sufficiently large to ensure that the sample distribution of a statistic (such as mean or proportion) is approximately normally distributed. Large samples are typically considered to have 30 or more observations.
