```python
## Reading the lung cancer dataset into pandas dataframe.

df=pd.read_csv("survey lung cancer.csv")
df.head()
```

```python
## shape of the given dataset.
df.shape
```

```python
## Baisc information about dataset
df.info()
```

- Total columns: 16
- Object Columns: 2
- Numeric Columns: 14
- Numeric Value Type: Most columns are discrete in nature. (values are 1 and 2)

```python
## Statstical Information about the dataset
df.describe().T
```

```python
## Checking Null values
df.isna().sum()
```

- No null values in this given data set.

```python
## Checking object type columns

print("Unique Values in Gender column: ",df['GENDER'].unique())
print("Unique Values in LUNG_CANCER column:",df['LUNG_CANCER'].unique())
```

```python
## COnverting the object type columns into numeric columns

df['GENDER']=df['GENDER'].map({'M':1, 'F':0})
df['LUNG_CANCER']=df['LUNG_CANCER'].map({'YES':1, 'NO':0})
```

```python
## checking the final df
## Now all the columns are same data types.
df.dtypes
```

```python
## Checking for the data balance or imbalance

df['LUNG_CANCER'].value_counts()
```

- Given data is imbalance
- Number of people have cancer 270
- Number of people not have cancer 39

# Graphical visualization

```python
plt.figure(figsize=(20,10))
sns.clustermap(data=df.corr(), annot=True, vmin=-1, vmax=1)
plt.show()
```

```python
x = df['LUNG_CANCER'].value_counts().index.tolist()
y = df['LUNG_CANCER'].value_counts().tolist()
plt.figure(figsize=(10,5))
sns.barplot( x=x, y=y)
plt.show()
```

# Checking the  distribution  of All features


```python

plt.figure(figsize=(30,30))
col_index=1

for col in df.columns:
  plt.subplot(4,4,col_index)  
  sns.histplot(df, x=df[col], kde=True)
  plt.title(f'Distribution of {col}.')
  plt.xlabel(col)
  col_index += 1
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
